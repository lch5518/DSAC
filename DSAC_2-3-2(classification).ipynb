##### 붓꽃(iris)

![iris](https://www.dropbox.com./s/jfep97cix2ajukk/ml-2.png?raw=1)

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

from sklearn import datasets

iris = datasets.load_iris()

#type(iris)
iris

X = iris.data
y = iris.target

X2 = X[:,:2]  # extract noly the first two columns

#type(X), type(y)

my_marker = ['*','o','v']
my_colors = ['r','b','g']

for i in range(3):
    xs = X2[:,0][y==i]
    ys = X2[:,1][y==i]
    plt.scatter(xs, ys, marker = my_marker[i], c = my_colors[i])

plt.title('viewing the dataset')
plt.legend(iris.target_names)
plt.xlabel('sepal length')
plt.ylabel('sepal width')
```

```python
# 두가지 군집으로 줄여보자
y2 = y.copy()
y2[y2==2]=1  # y2에서 2를 1로 변경

y,y2

X = iris.data
y2 = iris.target

X2 = X[:,:2]  # extract noly the first two columns

#type(X), type(y)

my_marker = ['*','o']
my_colors = ['r','b']

for i in range(2):
    xs = X2[:,0][y==i]
    ys = X2[:,1][y==i]
    plt.scatter(xs, ys, marker = my_marker[i], c = my_colors[i])
    
plt.title('viewing the dataset  : setosa - & non-setosa')
names = ['setosa','non-setosa']
plt.legend(names)
plt.xlabel('sepal length')
plt.ylabel('sepal width')
```

```python
from sklearn.linear_model import SGDClassifier
model = SGDClassifier()
model.fit(X2,y2)
model.coef_, model.intercept_    # coef_ : w, intercept_ : b 
```

```python
np.random.seed(42)
z = np.random.rand(10,2).round(2) * 10
model.predict(z)
```

## nonlinear classifier using decisiontree

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

from sklearn import datasets

iris = datasets.load_iris()

#type(iris)
iris

X = iris.data
y = iris.target

# 두가지 군집으로 줄여보자
X2 = X[:,:2]  # extract noly the first two columns
y2 = y.copy()
y2[y2==2]=1  # y2에서 2를 1로 변경


from sklearn.tree import DecisionTreeClassifier
dt_model = DecisionTreeClassifier()
dt_model.fit(X2,y2)
dt_model.classes_
```

```python
np.random.seed(42)
z = np.random.rand(10,2).round(2) * 10
dt_model.predict(z)
```
